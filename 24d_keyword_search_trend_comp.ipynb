## 24d - 검색어추이 데이터 가공

```python
# 아이폰14
data_i14=pd.read_csv('C:/herick/Projects/NCP/pythonCode/24d_i14_검색어추이.csv',
                     names=['date','count', 'day', 'model'])
data_i14.head()
```

```python
"""전처리"""
data_i14=data_i14.iloc[1:,:]# 첫번째 행 삭제
data_i14['count']=data_i14['count'].apply(lambda x: int(x) if x.isdigit() else np.nan) # count 변수를 obejct에서 int로 변경
data_i14['day']=data_i14['day'].apply(lambda x: int(x) if x.isdigit() else np.nan) # day 변수를 obejct에서 int로 변경
data_i14['date']=pd.to_datetime(data_i14['date'], errors='coerce') # 날짜 형식으로 변경
data_i14.reset_index(drop=True,inplace=True)
data_i14.head()
```
